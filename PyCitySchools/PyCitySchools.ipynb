```python
import pandas as pd
```

```python
#file paths to load
school_data_to_load="Resources/schools.csv"
student_data_to_load="Resources/students.csv"
```

```python
#read school and student data files and store into dataframes
school_data_df=pd.read_csv(school_data_to_load)
student_data_df=pd.read_csv(student_data_to_load)
```

```python
#merge the data into a single dataframe
school_data_complete_df = pd.merge(student_data_df,school_data_df, how="left", on=["school_name","school_name"])
school_data_complete_df.head()
```

```python
#District Summary
#Find the total number of schools
total_schools=len(school_data_complete_df["school_name"].value_counts())
total_schools
#Find the total number of students
total_students=len(school_data_complete_df["Student ID"].value_counts())
total_students
#Find total budget
total_budget=school_data_df["budget"].sum()
total_budget
#Average Math Score
average_math_score=student_data_df["math_score"].mean()
average_math_score
#Average Reading Score
average_reading_score=student_data_df["reading_score"].mean()
average_reading_score
#Percentage of Students with a passing math score
passing_math_score_df=student_data_df.loc[(student_data_df["math_score"]>=70)]
percent_passed_math=(len(passing_math_score_df["Student ID"].value_counts())/len(student_data_df["Student ID"].value_counts()))*100
percent_passed_math
#Percentage of Students with a passing english score
passing_reading_score_df=student_data_df.loc[(student_data_df["reading_score"]>=70)]
percent_passed_reading=(len(passing_reading_score_df["Student ID"].value_counts())/len(student_data_df["Student ID"].value_counts()))*100
percent_passed_reading
#Find percentage of students who passed overall, which means they passed the math and reading parts
passing_overall_df=student_data_df.loc[(student_data_df["reading_score"]>=70)&(student_data_df["math_score"]>=70)]
percent_passed_overall=(len(passing_overall_df["Student ID"].value_counts())/len(student_data_df["Student ID"].value_counts()))*100
percent_passed_overall
#Create a dataframe to hold the above results
district_summary_df=pd.DataFrame({
    "Total Schools":[total_schools],
    "Total Students":[total_students],
    "Total Budget":[total_budget],
    "Average Math Score": [average_math_score],
    "Average Reading Score":[average_reading_score],
    "% Passing Math": [percent_passed_math],
    "% Passing Reading":[percent_passed_reading],
    "Overall Passing":[percent_passed_overall]
})

district_summary_df["Total Students"]=district_summary_df["Total Students"].map("{:,}".format)
district_summary_df["Total Budget"]=district_summary_df["Total Budget"].map("${:,.2f}".format)
district_summary_df["Average Math Score"]=district_summary_df["Average Math Score"].map("{:.2f}".format)
district_summary_df["Average Reading Score"]=district_summary_df["Average Reading Score"].map("{:.2f}".format)
district_summary_df["% Passing Math"]=district_summary_df["% Passing Math"].map("{:.2f}%".format)
district_summary_df["% Passing Reading"]=district_summary_df["% Passing Reading"].map("{:.2f}%".format)
district_summary_df["Overall Passing"]=district_summary_df["Overall Passing"].map("{:.2f}%".format)

district_summary_df
```

```python
#Creating the school summary table
#set the index to name and the first column to type
school_type=school_data_df.set_index(["school_name"])["type"]
#Total Students per school 
total_students=school_data_complete_df.groupby("school_name").count()["Student ID"]
#Per School and Per Student Budget
per_school_budget = school_data_complete_df.groupby("school_name").mean()["budget"]
per_student_budget = per_school_budget/total_students
#Average Math Score of each school
average_math_by_school=student_data_df.groupby(["school_name"])["math_score"].mean()
#Average Reading Score of each school
average_reading_by_school=student_data_df.groupby(["school_name"])["reading_score"].mean()
#Percent passed per school math
students_passed_math=school_data_complete_df[school_data_complete_df["math_score"]>=70].groupby("school_name").count()["student_name"]
percent_passed_math=(students_passed_math/total_students)*100
#Percent passed per school reading
students_passed_reading=school_data_complete_df[school_data_complete_df["reading_score"]>=70].groupby("school_name").count()["student_name"]
percent_passed_reading=(students_passed_reading/total_students)*100
#Percent overall
students_passed_overall=school_data_complete_df[(school_data_complete_df["math_score"]>=70) & (school_data_complete_df["reading_score"]>=70)].groupby("school_name").count()["student_name"]
percent_passed_overall=(students_passed_overall/total_students)*100
#Create the data frame to show school summary
school_summary_df=pd.DataFrame({
    "School Types":school_type,
    "Total Students":total_students,
    "Total School Budget":per_school_budget,
    "Per Student Budget":per_student_budget,
    "Average Math Score":average_math_by_school,
    "Average Reading Score":average_reading_by_school,
    "% Passing Math":percent_passed_math,
    "% Passing Reading":percent_passed_reading,
    "% Passing Overall":percent_passed_overall
    })
#Define Columns for the Data Frame       
school_summary_df = school_summary_df[["School Types","Total Students","Total School Budget","Per Student Budget","Average Math Score","Average Reading Score","% Passing Math","% Passing Reading","% Passing Overall"]]
#adjust formatting using mapping
school_summary_df["Total School Budget"]=school_summary_df["Total School Budget"].map("${:,.2f}".format)
school_summary_df["Average Math Score"]=school_summary_df["Average Math Score"].map("{:.2f}".format)
school_summary_df["Average Reading Score"]=school_summary_df["Average Reading Score"].map("{:.2f}".format)
school_summary_df["% Passing Math"]=school_summary_df["% Passing Math"].map("{:.2f}%".format)
school_summary_df["% Passing Reading"]=school_summary_df["% Passing Reading"].map("{:.2f}%".format)
school_summary_df["% Passing Overall"]=school_summary_df["% Passing Overall"].map("{:.2f}%".format)
#show dataframe
school_summary_df   
```

```python
#Top  5 Performing Schools by Overall Passing
# set the index and first column
school_type=school_data_df.set_index(["school_name"])["type"]
# total students per school
total_students=school_data_complete_df.groupby("school_name").count()["Student ID"]
# Per School Budget
per_school_budget = school_data_complete_df.groupby("school_name").mean()["budget"]
per_student_budget = per_school_budget/total_students
#Average Math Score
average_math_by_school=student_data_df.groupby(["school_name"])["math_score"].mean()
#Average Reading Score
average_reading_by_school=student_data_df.groupby(["school_name"])["reading_score"].mean()
#Percent passed per school math
students_passed_math=school_data_complete_df[school_data_complete_df["math_score"]>=70].groupby("school_name").count()["student_name"]
percent_passed_math=(students_passed_math/total_students)*100
#Percent passed per school reading
students_passed_reading=school_data_complete_df[school_data_complete_df["reading_score"]>=70].groupby("school_name").count()["student_name"]
percent_passed_reading=(students_passed_reading/total_students)*100
#Percent overall
students_passed_overall=school_data_complete_df[(school_data_complete_df["math_score"]>=70) & (school_data_complete_df["reading_score"]>=70)].groupby("school_name").count()["student_name"]
percent_passed_overall=(students_passed_overall/total_students)*100
#Create the data frame to show school summary
top_performing_schools_df=pd.DataFrame({
    "School Types":school_type,
    "Total Students":total_students,
    "Total School Budget":per_school_budget,
    "Per Student Budget":per_student_budget,
    "Average Math Score":average_math_by_school,
    "Average Reading Score":average_reading_by_school,
    "% Passing Math":percent_passed_math,
    "% Passing Reading":percent_passed_reading,
    "% Passing Overall":percent_passed_overall
    })
#Define Columns for the Data Frame       
top_performing_schools_df = top_performing_schools_df[["School Types","Total Students","Total School Budget","Per Student Budget","Average Math Score","Average Reading Score","% Passing Math","% Passing Reading","% Passing Overall"]]
#Apply neccessary sorting
top_performing_schools_df= top_performing_schools_df.sort_values("% Passing Overall", ascending=False)
#adjust formatting using mapping
top_performing_schools_df ["Total Students"]=top_performing_schools_df ["Total Students"].map("{:,}".format)
top_performing_schools_df ["Total School Budget"]=top_performing_schools_df ["Total School Budget"].map("${:,.2f}".format)
top_performing_schools_df ["Average Math Score"]=top_performing_schools_df ["Average Math Score"].map("{:.2f}".format)
top_performing_schools_df ["Average Reading Score"]=top_performing_schools_df ["Average Reading Score"].map("{:.2f}".format)
top_performing_schools_df ["% Passing Math"]=top_performing_schools_df ["% Passing Math"].map("{:.2f}%".format)
top_performing_schools_df ["% Passing Reading"]=top_performing_schools_df ["% Passing Reading"].map("{:.2f}%".format)
top_performing_schools_df ["% Passing Overall"]=top_performing_schools_df ["% Passing Overall"].map("{:.2f}%".format)
#show dataframe
top_performing_schools_df.head(5) 
```

```python
#Bottom 5 Performing Schools
# take previous dataframe but show in ascending order
top_performing_schools_df=top_performing_schools_df.sort_values("% Passing Overall")
top_performing_schools_df.head(5)
```

```python
#Math Scores by grade
#set the index
school_type=school_data_df.set_index(["school_name"])["type"]
#for 9th grade
freshmen_math=student_data_df.loc[student_data_df["grade"]=="9th"].groupby("school_name")["math_score"].mean()
#for 10th grade
sophomore_math=student_data_df.loc[student_data_df["grade"]=="10th"].groupby("school_name")["math_score"].mean()
#for 11th grade
junior_math=student_data_df.loc[student_data_df["grade"]=="11th"].groupby("school_name")["math_score"].mean()
#for 12th grade
senior_math=student_data_df.loc[student_data_df["grade"]=="12th"].groupby("school_name")["math_score"].mean()
#Create the dataframe
math_score_by_grade_df=pd.DataFrame({"School Type":school_type,
              "9th":freshmen_math,
              "10th":sophomore_math,
              "11th":junior_math,
              "12th":senior_math 
             })
math_score_by_grade_df=math_score_by_grade_df[["9th","10th","11th","12th"]]
#format the columns
math_score_by_grade_df["9th"]=math_score_by_grade_df["9th"].map("{:.2f}".format)
math_score_by_grade_df["10th"]=math_score_by_grade_df["10th"].map("{:.2f}".format)
math_score_by_grade_df["11th"]=math_score_by_grade_df["11th"].map("{:.2f}".format)
math_score_by_grade_df["12th"]=math_score_by_grade_df["12th"].map("{:.2f}".format)
#show dataframe
math_score_by_grade_df
```

```python
#Reading Scores by grade
#set the index
school_type=school_data_df.set_index(["school_name"])["type"]
#for 9th grade
freshmen_reading=student_data_df.loc[student_data_df["grade"]=="9th"].groupby("school_name")["reading_score"].mean()
#for 10th grade
sophomore_reading=student_data_df.loc[student_data_df["grade"]=="10th"].groupby("school_name")["reading_score"].mean()
#for 11th grade
junior_reading=student_data_df.loc[student_data_df["grade"]=="11th"].groupby("school_name")["reading_score"].mean()
#for 12th grade
senior_reading=student_data_df.loc[student_data_df["grade"]=="12th"].groupby("school_name")["reading_score"].mean()
#Create the dataframe
reading_score_by_grade_df=pd.DataFrame({"School Type":school_type,
              "9th":freshmen_reading,
              "10th":sophomore_reading,
              "11th":junior_reading,
              "12th":senior_reading 
             })
reading_score_by_grade_df=reading_score_by_grade_df[["9th","10th","11th","12th"]]
#format the columns
reading_score_by_grade_df["9th"]=reading_score_by_grade_df["9th"].map("{:.2f}".format)
reading_score_by_grade_df["10th"]=reading_score_by_grade_df["10th"].map("{:.2f}".format)
reading_score_by_grade_df["11th"]=reading_score_by_grade_df["11th"].map("{:.2f}".format)
reading_score_by_grade_df["12th"]=reading_score_by_grade_df["12th"].map("{:.2f}".format)
#show dataframe
reading_score_by_grade_df
```

```python
#Scores by School Spending
schools_spending_df=school_summary_df[["Per Student Budget","Average Math Score","Average Reading Score","% Passing Math","% Passing Reading","% Passing Overall"]]
schools_spending_df
#Bins
bins=[575,600,620,640,660]
group_names=["<$600","$600-620","$620-640","$640-660"]
#new data frame
schools_spending_df["Spending Range Per Student"] = pd.cut(schools_spending_df["Per Student Budget"], bins, labels=group_names, include_lowest=True)
schools_spending_df
#delete unneccessary row
del schools_spending_df["Per Student Budget"]
schools_spending_df
#groupby spending range
schools_spending_df = schools_spending_df.groupby("Spending Range Per Student")
schools_spending_df.max()
```

```python
#Scores by School Size
school_size_df=school_summary_df[["Total Students","Average Math Score","Average Reading Score","% Passing Math","% Passing Reading","% Passing Overall"]]
school_size_df
#bins
bins=[400,1000,2000,5000]
group_names=["small(<1000)","medium(1000-2000)","large(2000-5000)"]
#adjust data frame with new column
school_size_df["School Size"] = pd.cut(school_size_df["Total Students"], bins, labels=group_names, include_lowest=True)
school_size_df
#delete unneccessary row
del school_size_df["Total Students"]
#groupby dataframe
school_size_df = school_size_df.groupby("School Size")
school_size_df.max()
```
